The Gapminder chart above illustrates the relationship between the quantity of fertilizer used by a country (plotted on the x-axis) and the total output produced (plotted on the y-axis). As a third variable, we have selected the quantity of agricultural land, which represents the amount of pastures used for livestock grazing and croplands available in a country.

At first glance, both charts shows a strong positive correlation between the amount of fertilizer used by a country and the output they produce. This result is expected, as fertilizers are additives designed to increase the productivity of crop yields. Additionally, the Gapminder charts indicate that an increase in the amount of agricultural land available has a positive effect on output, as the area of data points generally increases with an increase in output quantity. This rule, however, is not as strong, as we can observe several smaller countries generating greater outputs through the use of more fertilizers.

When comparing 1980 with 2019, it is evident that the circles began to overlap in 2019 and were slightly larger than the previous year. This suggests a greater amount of pastures used in overall production. However, this change is relatively small when considering the time span of 40 years. One might expect more significant changes in the data during such a long period. Overllaping of the data may suggest that there is a limit to how much output can be produced even with increased fertilizer use and available agricultural land.
